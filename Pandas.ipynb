Write a function that takes start and end of a range returns a Pandas series object containing numbers within that range.

In case the user does not pass start or end or both they should default to 1 and 10 respectively. 

eg. 

range_series() -> Should Return a pandas series from 1 to 10

range_series(5) -> Should Return a pandas series from 5 to 10 

range_series(5, 10) -> Should Return a pandas series from 5 to 15

```python
def range_series():
    a = input("Enter the start value: ")
    b = input("Enter the end value: ")
    if a == '' and b == '':
        df1 = pd.Series(np.arange(1,11))
    elif a == '' or b == '':
        if a == '':
            r = int(b)
        else:
            r = int(a)
        df1 = pd.Series(np.arange(r,r+6))
    else:
        arr1 = np.arange(int(a),int(b)+6)
        df1 = pd.Series(arr1)
    return df1
x = range_series()
x
```

Create a function that concatenates two dataframes. Use previously created function to create two dataframes and pass them as parameters Make sure that the indexes are reset before returning

```python
def concat(x,y):
    z = pd.concat([x,y],ignore_index=True,axis = 1)
    return z
concat(x,y)
```

Write code to load data from cars.csv into a dataframe and print its details. Details like: 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'.

```python
df3 = pd.read_csv("mtcars.csv")
df3.describe()
```

```python
df3.head(3)
```

```python
df3.corr()
```

```python
def max_corr_column(clm_name):
    corr_tab = df3.corr()
    x = corr_tab.sort_values(by = clm_name, ascending=True)
    y = x[x[clm_name][-2] == x[clm_name]].index
    return y

max_corr_column('mpg')
```
