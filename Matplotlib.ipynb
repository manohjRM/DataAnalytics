```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfcc = pd.read_csv("customer_churn.csv")
dfcc.head()
```

```python
contract = dfcc['Contract'].value_counts()
x = contract.index
y = contract.values
plt.bar(x,y,color = 'orange')
plt.xlabel('Contract Type of customer')
plt.ylabel('Count')
plt.title('Distribution of Contract')
```

```python
plt.hist(dfcc['MonthlyCharges'],histtype='stepfilled',color='forestgreen')
plt.xlabel('Monthly Charges Incurred')
plt.ylabel("Count")
plt.title('Distribution of Montly Charges')
```

```python
plt.scatter(x = dfcc['tenure'],y = dfcc['TotalCharges'])
```

```python
dfcc.boxplot(column = 'MonthlyCharges',by = ['PaymentMethod'],color = 'olive')
plt.xlabel('Payment Method of customer')
plt.ylabel("Monthly Charges Incurred")
plt.title('Monthly Charges vs Payment Method')
```
